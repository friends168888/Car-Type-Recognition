```python
from keras.preprocessing.image import img_to_array
from keras.models import load_model
import numpy as np
import argparse
import imutils
import cv2
from imutils import paths
import os
import shutil
```

```python
path = 'C:\\Users\\Big data\\PycharmProjects\\ML\\dataset'
imagePaths = sorted(list(paths.list_images(path)))
model = load_model('classifier0802.h5')
```

```python
for imagePath in imagePaths:
    # load the image
    print(imagePath)
    image = cv2.imread(imagePath)
#     orig = image.copy()

    # pre-process the image for classification
    print(imagePath)
    try:
        image = cv2.resize(image, (128, 128))
    except:
        shutil.move(imagePath, 'C:\\Users\\Big data\\PycharmProjects\\ML\\c3')
        continue
    image = image.astype("float") / 255.0
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    result = model.predict(image)[0]
    proba = np.max(result)
    print(imagePath)
    print(proba)
    if proba < 0.95:
        try:
            shutil.move(imagePath, 'C:\\Users\\Big data\\PycharmProjects\\ML\\c1')
        except shutil.Error:
            try:
                shutil.move(imagePath, 'C:\\Users\\Big data\\PycharmProjects\\ML\\c2')
            except shutil.Error:
                try:
                    shutil.move(imagePath, 'C:\\Users\\Big data\\PycharmProjects\\ML\\c3')
                except shutil.Error:
                    os.remove(imagePath)
```

```python
import os  # 

def CEF(path):
    """
    CLean empty files, 
    :param path: 
    :return: None
    """
    files = os.listdir(path)  # ()
    for file in files:
        print('Traversal at', file)
        if os.path.isdir(file):  # 
            if not os.listdir(file):  # 
                os.rmdir(file)  # 
        elif os.path.isfile(file):  # 
            if os.path.getsize(file) == 0:  # 0
                os.remove(file)  # 
    print(path, 'Dispose over!')

if __name__ == "__main__":  # 
    path = 'C:\\Users\\Big data\\PycharmProjects\\ML\\model0731'  # 
    CEF(path)
```
